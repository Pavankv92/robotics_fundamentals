### Docker Fundamentals

#### Image

- **Definition**: Immutable templates containing pre-installed files, binaries, libraries, and dependencies needed to run a program. These are blueprints for instantiating containers. Note that a Docker image is not a full operating system.
- **How to obtain an image**:
  - From Docker Hub (public registry).
  - Create one using a Dockerfile (a list of instructions). The Docker engine parses the Dockerfile and builds a new image.
  - Share a compressed version of an existing Docker image.

#### Container

- **Definition**: A running instance of a Docker image. Once running, you can install libraries or software inside it just like a real OS. Containers are temporary and can be easily created or deleted. Changes made to a container are lost when it's deleted.
- **Key characteristics**:
  - Containers run on top of the host PC's OS.
  - They rely on the host's kernel rather than virtualizing it like a virtual machine.
  - Can access host PC's GPUs directly.

### Docker Hub

- **What is Docker Hub**: The official public container registry provided by Docker.
  - Hosts millions of publicly available images (official, certified, community).
  - Supports private repositories with a Docker Hub account (free tier for small teams).
- **Key concepts**:
  - **Repository**: Named collection of images (e.g., `library/python`).
  - **Tag**: Version identifier for images in the repository (e.g., `latest`, `3.11-slim`).
  - **Image manifest**: Metadata defining layers and configuration.
- **Common Docker Hub workflows**:
  - Search images: `docker search nginx`
  - Pull an image: `docker pull ubuntu:24.04`
  - Inspect image metadata: `docker image inspect ubuntu:24.04`
  - List local images: `docker images`
  - Remove image: `docker rmi ubuntu:24.04`
- **Push your own image**:
  1. Build: `docker build -t yourusername/robotics_fundamentals:latest .`
  2. Login: `docker login` (enter Docker Hub credentials)
  3. Push: `docker push yourusername/robotics_fundamentals:latest`
  4. Pull from another machine: `docker pull yourusername/robotics_fundamentals:latest`
- **Best practices**:
  - Use small base images (e.g., `alpine`, `python:3.12-slim`).
  - Pin explicit tags in production (`python:3.12.0-slim` instead of `latest`).
  - Minimize layers by combining commands where appropriate.
  - Keep secrets out of Dockerfiles; use build-time or runtime secrets.
  - Use automated builds / GitHub integration to publish images with CI.
- **Extras**:
  - Connect with GitHub/GitLab for automated builds on push.
  - Enable Docker Content Trust for image signing and verification.
  - Monitor security with `docker scan` and image vulnerability scanning tools.


### More on Docker Images

- **Layered Architecture**: Docker images are built from multiple layers, each representing a set of changes (e.g., base layer → Ubuntu 24.04 → ROS2 → Python). Each layer is immutable and has a unique hash for efficient storage and caching.
- **Layer Sharing**: If multiple images share the same layer (e.g., Ubuntu 24.04), Docker stores it only once on disk, saving space. When pulling images from Docker Hub, shared layers aren't re-downloaded if already present locally.
- **Image as Base Layer**: An entire image can serve as the base layer for building new images, allowing for inheritance and modularity. Every layer and the complete image receives a unique ID.
- **Container Instantiation**: Multiple containers can be created from a single image. Each container gets a unique hash ID and can be assigned user-friendly names or randomized names to avoid conflicts.


### Container Lifecycle Management

- **Create a container**: `docker create <image>` → Container state: `created` (container exists but not running).
- **Start the container**: `docker start <container>` → Container state: `running` (processes inside are active).
- **Pause/Resume the container**: `docker pause <container>` / `docker unpause <container>` → Container state: `paused` / `running` (freezes/unfreezes all processes without stopping).
- **Stop the container**: `docker stop <container>` → Container state: `stopped` (gracefully shuts down, still exists on disk).
- **Kill the container**: `docker kill <container>` → Container state: `stopped` (forcefully terminates, for unresponsive containers).
- **Restart the container**: `docker restart <container>` → Container state: `running` (stops and starts the container).
- **Delete the container**: `docker rm <container>` → Container state: `deleted` (permanently removed from filesystem; data lost unless persisted).
- **Additional commands**:
  - List containers: `docker ps` (running), `docker ps -a` (all states).
  - Inspect container: `docker inspect <container>` (detailed metadata).
  - Logs: `docker logs <container>` (view stdout/stderr).
  - Execute commands: `docker exec -it <container> <command>` (run inside running container).
- **Best practices**:
  - Use volumes or bind mounts for persistent data.
  - Clean up stopped containers regularly: `docker container prune`.
  - Monitor resource usage: `docker stats <container>`.
  - Use restart policies (e.g., `--restart=always`) for production resilience.

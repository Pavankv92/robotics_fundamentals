### Environment variables

- **Print an environment variable:**

  ```bash
  echo "$HOME"
  ```

- **List all environment variables:**

  ```bash
  printenv
  ```

- **Set a new variable (valid only for the current shell session):**

  ```bash
  export MY_VAR="Hello"
  ```

- **Make the variable permanent (add it to your shell startup file):**

  ```bash
  echo 'export MY_VAR="Hello"' >> ~/.bashrc
  ```

- **Unset an environment variable (session only):**
  ```bash
  unset MY_VAR
  ```

### Aliases

Aliases allow you to create shortcuts for frequently used commands. An alias can be a simple command or a complex sequence of commands.

- **Create an alias (valid only for the current shell session):**

  ```bash
  alias ll='ls -la'
  alias backup='cp -r ~/data ~/backups/$(date +%F)'
  ```

- **Make an alias permanent (add it to your shell startup file):**

  ```bash
  echo "alias ll='ls -la'" >> ~/.bashrc
  echo "alias backup='cp -r ~/data ~/backups/$(date +%F)'" >> ~/.bashrc
  ```

- **Remove an alias (session only):**
  ```bash
  unalias ll
  ```


### Bash Scripting


#### Scheduling Scripts with CRON

- **Basic usage:** To schedule a bash script to run at a specific time, use CRON (a task scheduler). Edit the crontab with `crontab -e` and add lines like:
  ```bash
  # Run script every day at 2 AM
  0 2 * * * /path/to/script.sh
  ```

#### Running Programs in the Background with systemd

- **Overview:** To run a program continuously in the background and restart it automatically (e.g., on reboot or crash), use systemd (an init system and service manager in modern Linux systems like Ubuntu).

- **Management:** Systemd manages daemons (processes that run in the background, e.g., an SSH server called `sshd`). Use the `systemctl` tool to manage daemons: start, stop, check status, etc.

- **Example - Check status:**

  ```bash
  sudo systemctl status ssh
  ```

- **Creating custom services:** To create a custom service, create a unit file in `/etc/systemd/system/` (e.g., `my-service.service`) and use `systemctl` to enable and start it.

- **Comparison:** CRON is great for scheduling, but it doesn't supervise running processes. Systemd handles both launching and supervising daemons.


#### Basic Bash Script Example

Create a simple script to list files in the current directory:

```bash
#!/bin/bash
echo "Current directory: $(pwd)"
echo "Files:"
ls -la
```

- **Make it executable:**

  ```bash
  chmod +x script.sh
  ```

#### Useful Tips

- Always start scripts with `#!/bin/bash` (shebang).
- Use variables: `MY_VAR="value"` (no spaces around `=`).
- Conditionals: `if [ "$VAR" == "value" ]; then ... fi`
- Loops: `for file in *.txt; do echo "$file"; done`


#### Advanced Bash Script Example: Directory Backup

Create a bash script to make a backup of a given directory with a date-stamped folder name.

Make it executable and run:

```bash
chmod +x bash_script_example.sh
./bash_script_example.sh
```


```bash
#!/usr/bin/env bash

# --- create a bash script to make a backup of given dir with a date as a folder name ---

# create env variables for src and backup dirs

SRC_DIR="$HOME"/data
BACKUP_DIR="$HOME/backups/$(date +%F)"

backup () {

    local src="$1" # catch the first argument passed
    local dest="$2" # catch the second argument passed

    if [[ -d "$src" ]]; then
        mkdir -p "$dest"
        cp -r "$src"/* "$dest"
        echo "Backup of $src folder into --> $dest"

    else
        echo "Source dir $src doesn't exist!"
    fi

}

#### create SRC_DIR and run the script

backup $SRC_DIR $BACKUP_DIR
```

### Using systemd to add, run a background service

#### Creating a simple logs.sh bash script

Create a simple logs.sh script:

```bash
#!/usr/bin/env bash

LOGFILE="/tmp/service.log"

while true; do
    echo "Logs saved at $(date)" >> $LOGFILE
    sleep 3
done
```

- **Make the file executable:**
  ```bash
  chmod +x logs.sh
  ```

#### Creating a systemd service file

Create a `logs.service` file in `/etc/systemd/system`:

```bash
[Unit]
Description=Log file service

[Service]
ExecStart=/home/robot/robotics_fundamentals/scripts/linux/logs.sh
Restart=always
User=robot

[Install]
WantedBy=multi-user.target
```

#### Managing the service

Run these commands in the terminal:

- **Restart the daemon:**

  ```bash
  sudo systemctl daemon-reload
  ```

- **Enable the service:**

  ```bash
  sudo systemctl enable logs.service
  ```

- **Start the service:**

  ```bash
  sudo systemctl start logs.service
  ```

- **Monitor the log file:**
  ```bash
  tail -f /tmp/service.log
  ```

### Searching

#### Reverse Search (Ctrl + R)

- **Description:** In the bash shell, Ctrl + R allows you to perform a reverse search through your command history. This is useful for quickly finding and reusing previously entered commands.

- **How to use:**
  - Press `Ctrl + R` in the terminal.
  - Start typing part of the command you want to find.
  - The shell will show the most recent matching command.
  - Press `Ctrl + R` again to cycle through older matches.
  - Press `Enter` to execute the command, or `Ctrl + C` to cancel.

- **Example:**
  - If you've previously run `ls -la /home/user`, pressing `Ctrl + R` and typing "ls" will bring up that command.

#### Grep

- **Description:** `grep` is a powerful command-line tool for searching text patterns in files. It supports regular expressions and various options for filtering and formatting output.

- **Basic syntax:** `grep [options] pattern [file(s)]`

- **Examples:**
  - Search for a specific word in a file:
    ```bash
    grep "error" logfile.txt
    ```
  - Case-insensitive search:
    ```bash
    grep -i "Error" logfile.txt
    ```
  - Search recursively in a directory:
    ```bash
    grep -r "pattern" /path/to/directory
    ```
  - Count occurrences:
    ```bash
    grep -c "word" file.txt
    ```
  - Show line numbers:
    ```bash
    grep -n "pattern" file.txt
    ```

#### Awk

- **Description:** `awk` is a versatile text processing language and command-line tool. It's particularly useful for manipulating structured text data, such as CSV files or log files with columns.

- **Basic syntax:** `awk 'pattern { action }' file(s)`

- **Examples:**
  - Print the first column of a file:
    ```bash
    awk '{print $1}' data.txt
    ```
  - Print lines where the second column is greater than 100:
    ```bash
    awk '$2 > 100 {print}' data.txt
    ```
  - Sum all values in the third column:
    ```bash
    awk '{sum += $3} END {print "Total:", sum}' data.txt
    ```
  - Process CSV files (assuming comma-separated):
    ```bash
    awk -F',' '{print $1 "," $3}' file.csv
    ```
  - Print lines matching a pattern:
    ```bash
    awk '/ERROR/ {print NR ": " $0}' logfile.txt
    ```


### Tmux

#### Why Use Tmux?

Tmux (Terminal Multiplexer) is a powerful tool that allows you to:

- **Multiplexing:** Run multiple terminals in a single window using panes and windows.
- **Session Management:** Create and manage multiple sessions that persist even if you disconnect from SSH.
- **Productivity:** Switch between multiple projects and terminal sessions without opening new windows.
- **Remote Sessions:** Reattach to running sessions, which is especially useful for remote work and long-running processes.
- **Scripting and Automation:** Automate terminal operations with scripting.

**Key advantages:**

- Keep processes running even if your connection drops.
- Easy window and pane management with keyboard shortcuts.
- Share terminals between multiple users.
- Work with multiple projects simultaneously in organized layouts.

#### How to Use Tmux

**Installation:**

```bash
sudo apt install tmux
```

**Basic Commands:**

- **Start a new session:**

  ```bash
  tmux new-session -s myproject
  # or shorthand
  tmux new -s myproject
  ```

- **List all sessions:**

  ```bash
  tmux list-sessions
  # or
  tmux ls
  ```

- **Attach to an existing session:**

  ```bash
  tmux attach-session -t myproject
  # or
  tmux attach -t myproject
  ```

- **Detach from a session (inside tmux):**

  ```
  Ctrl + B, then D
  ```

- **Kill a session:**
  ```bash
  tmux kill-session -t myproject
  ```

**Working with Windows and Panes (inside tmux):**

- **Create a new window:**

  ```
  Ctrl + B, then C
  ```

- **Switch to next window:**

  ```
  Ctrl + B, then N
  ```

- **Switch to previous window:**

  ```
  Ctrl + B, then P
  ```

- **List all windows:**

  ```
  Ctrl + B, then W
  ```

- **Rename current window:**

  ```
  Ctrl + B, then ,
  ```

- **Split pane vertically (left-right):**

  ```
  Ctrl + B, then %
  ```

- **Split pane horizontally (top-bottom):**

  ```
  Ctrl + B, then "
  ```

- **Navigate between panes:**

  ```
  Ctrl + B, then Arrow Keys
  ```

- **Close current pane:**
  ```
  Ctrl + B, then X
  ```

**Practical Example:**

```bash
# Start a new session called 'dev'
tmux new -s dev

# Inside the session, split the pane
# Ctrl + B, then %

# In the left pane, run your development server
python app.py

# Switch to the right pane (Ctrl + B, Arrow Right) and edit files
nano config.py

# Create a new window for monitoring (Ctrl + B, then C)
# Run a monitoring command
tail -f logs.txt

# Detach from the session to keep it running (Ctrl + B, then D)

# Later, reattach to your work
tmux attach -t dev
```

**Advanced: Configuration**

Create a `.tmux.conf` file in your home directory to customize Tmux:

```bash
# Set a more convenient prefix key (Ctrl + A instead of Ctrl + B)
set -g prefix C-a
unbind C-b

# Enable mouse support
set -g mouse on

# Set terminal colors
set -g default-terminal "screen-256color"

# Increase scrollback history
set -g history-limit 10000

# Reload configuration with Ctrl + A, then R
bind r source-file ~/.tmux.conf \; display "Config reloaded!"
```

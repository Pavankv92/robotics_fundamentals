# Probability


Probability is a number between zero and one that measures how likely something is to happen [0,1].

### Notation

- X = Discrete random variable
- x = Continuous random variable
- P(A) = Probability of a event A occuring
- $P(A^-)$ = 1 - P(A) = Probability of a event A NOT occuring

### Rules of Probability

1.  $0 \le P(A) \le 1$

2.  $\sum_i P(A_i) = 1$

#### Sum Rule

Sum rule is generally concerned with with the probability of more than one mutually exclusive event happening. It tells us the probability of either event P(A or B) happening.

- Mutually exclusive event: two events A and B are said to mutually exclusive if they can't both happen at the same time. Either A or B happens. For example, a coin flip is either heads or tails; it can't be both. For a mutually exclusive events, if A happens, B is automatically excluded $P(B)= 0$ and vice versa.

- Independent: Probablity of two events are unrelated, probability of A is uneffected by wheather event B happend

$P(A \ or B) = P(A \cup B) = P(A) + P(B)$ for mutually exclusive events.

Example: For a unbiased die, proability of rolling four or five is $\frac{1}{6}$ + $\frac{1}{6}$ = $\frac{1}{3}$

#### Product Rule

The sum rule tells us the probability of events A or B happening. Product rule tells us probablity of event A and B happening. \
$P(A \ and \ B) = P(A \cap B) = P(A)P(B)$ for independent events.

for mutually exclusive events, its clear that $P(A)P(B) = 0 $

Example: Assume, 80% people in the world have brown eyes and 50% population is female. What is the probability of a randomly selected person being a female with brown eyes.

P(female, brown-eyed) = P(female) P(brown-eyed) = 0.5(0.8) = 0.40

#### Sum Rule (General)

Many events are not mutually exclusive. In that case, sum rule becomes

$P(A \ or B) = P(A \cup B) = P(A) + P(B) - P(A \ and \ B)$

Example: An archaeologist has discovered a dmall cache of 20 ancient coins. 12 coins are Roman and 8 are Greek. 6 of the Roman and 3 of the Greek are silver and remaining are bronze. What'd the probability of selecting the a silver or Roman coin from the cache ?

$P(silver \ or \ Roman) =  P(Silver) + P(Roman) =$ $\frac{9}{20}$ + $\frac{12}{20}$ = $\frac{21}{20} = 1.05 $ (Wrong!)

Roman coin is counted twice, once in silver and Roman. Use the sum rule (general)

$P(silver \ or \ Roman) =  P(Silver) + P(Roman) - P(Silver \ and \ Roman) =$ $\frac{9}{20}$ + $\frac{12}{20}$ - $\frac{6}{20}$= $\frac{15}{20} = 0.75$

### Conditional Probability

Consider a bag of marbles: 8 red and 2 blue. If we pick a marble at randomly from the bag, we have $\frac{2}{10}$ chance of picking a blue marble. Say we pick the blue marble, play with it and put it back in the same bag, shake it, and now pull our another marble. What's the chance of picking a blue marbke a second time ? it is still $\frac{2}{10}$.

If the fact that event A happened (here, picking a blue marble and putting it back in the bag) hasn't affected the probability of the future event B, the two are independent events. Our chance of picking a blue marble second time is in no way affected by the fact that we previously chose a blue marble.

Now consider an alternate scenario. We still have bag with 8 red and 2 blue marbles. We pick a marble, say red this time - and we keep this red marble aside. Now, we pick another marble from the bag. What's is the probability of picking another red marble? Here, things have changed. There are 9 marbles now and 7 of them red. So our chance of picking a red marble is $\frac{7}{9}$. The possibility of choosing a red marble were initially $\frac{8}{10}$, the fact that event A happened, has altered the probablity of the second event. The events are no longer independent. Notationally we write P(B|A) to mean the probability of event B given event A has happened. This is a conditional probability becuase it's conditional on event A happening.

Product Rule (Conditional) $P(A \ and \ B) = P(B|A)P(A)$. Probability of the 2 events both happening is the product of the probability of the one given the other has happened and the probability of the other. For our above scenario, $P(A \ and \ B) = P(B|A)P(A)$ = $\frac{7}{9}$ $\frac{8}{10}$ = 0.6222

- If the two events are mutually exclusive, $P(B|A)$ = $P(A|B)$ = 0.
- If two events are independent, then $P(B|A)$ = $P(B)$ and $P(A|B)$ = $P(A)$.
- Typically, $P(B|A)$ $\ne$ $P(A|B)$

### Total Probability

If our sample space is separated into disjoint regions, $B_i(B_1, B_2, etc.)$ so that the totality of th sample space is covered by the collection of $B_i$'s and the $B_i$'s don't overlap, we can calculate the probability of an event over all the partitions as the following:

$P(A)$ = $\sum_i P(A|B_i)P(B_i)$

Example: You have three cities, Kish, Kesh and Kuara, and their popolations are 2000, 1000 and 3000 respectively. Additionaly the percentages of people with blue eyes in these cities are 12%, 3% and 21% respectively. What's the probability that a randomly selected person from among the cities had blue eyes? The cities population affect things, as the probability of having blue eyes varies by the city and the cities vary in population. To find the P(blue), we use the total probability:

$ P(blue) = P(blue|Kish)P(Kish) + P(blue|Kesh)P(Kesh) + P(blue|Kuara)P(Kuara)$

$P(blue) = 0.12 \frac{2000}{6000} + 0.03 \frac{1000}{6000} + 0.21 \frac{3000}{6000} = 0.15$

### Joint & Marginal Probability

- The joint probabilities of two variables, $P(X=x, Y=y)$, is the probability that random variable $X$ will have a value $x$ at the same time random variable $Y$ is $y$.
  - A joint probability is the probability of multiple conditions being true at the same time, which is "and"
- The marginal probability is what we get when we calculate the probability of one or more of those conditions without caring for the value of others.
  - The probability of a subset of the random variables in the "and"

#### Joint probability table

Joint and marginal probabilities can be best understood via joint probability table. Pretend we survey 1000 people. We can count the number of people who are male and color-blind, female color-blind, male not color-blind, female Not color-blind. We arrange this data in a contiengency table below. Sum of the final row or column by necessity sums to the 1000.

|        | Color-blind | Not Color-blind |      |
| ------ | ----------- | --------------- | ---- |
| Male   | 42          | 456             | 498  |
| Female | 3           | 499             | 502  |
|        | 45          | 955             | 1000 |

We can turn the contingency table into a joint probability table simply by dividing by 1000.

|        | Color-blind | Not Color-blind |       |
| ------ | ----------- | --------------- | ----- |
| Male   | 0.042       | 0.456           | 0.498 |
| Female | 0.003       | 0.499           | 0.502 |
|        | 0.045       | 0.955           | 1.000 |

Now, we can look up probability values from the table.

$P(sex=male, color-blind= yes) = 0.042$

$P(sex=female, color-blind= no) = 0.499$

What if we wanted to know the probability of being color-blind regardless of sex ? For that we sum the probabilities along the color-blind column and see that it is 0.045. Likewise, summing along the rows gives us an estimated probability of being female is 0.498.

Calculating the probability of being color-blind or female from the joint probability table is calculating a marginal probability. In the first case, we summed along the column to remove the effect of sex, whereas in the second case, we summned along the row to remove the effect of color-blindness.

Mathematically, we get the marginal probabilities by summing across the vasitables we don't want.

$P(X=x)=\sum_{i} P(X=x, Y=y_{i})$ sum along $y_i$'s

$P(Y=y)=\sum_{i} P(Y=y, X=x_{i})$ sum along $x_i$'s

for example,

$P(Y=color-blind)= P(X=male, Y=color-blind) + P(X=female, Y=color-blind) $

### Probability Distributions

Probability distributions van be thought of as a function that generates values on demand. The values generated are random - we dont know which one will appear - but the likelyhood of any value appearing follows a general form. There are discrete probability distribution (PMF:probability mass function) and continuous probability distributions (PDF:probability density function). The figure below shows some of the most important continuous probability distributions, Uniform, Gaussian(normal), Gamma and Beta.


```python
%pip install matplotlib
import numpy as np
import matplotlib.pylab as plt

N = 10000000
B = 100
x = np.arange(B) / B

#  uniform
t = np.random.random(N)
u = np.histogram(t, bins=B)[0]
u = u / u.sum()

#  normal
t = np.random.normal(0, 1, size=N)
n = np.histogram(t, bins=B)[0]
n = n / n.sum()

#  gamma
t = np.random.gamma(5.0, size=N)
g = np.histogram(t, bins=B)[0]
g = g / g.sum()

#  beta
t = np.random.beta(5, 2, size=N)
b = np.histogram(t, bins=B)[0]
b = b / b.sum()

plt.plot(x, u, color="k", linestyle="solid", label="Uniform")
plt.plot(x, n, color="k", linestyle="dotted", label="Normal")
plt.plot(x, g, color="k", linestyle="dashed", label="Gamma")
plt.plot(x, b, color="k", linestyle="dashdot", label="Beta")
plt.ylabel("Probability")
# plt.tight_layout(pad=0, w_pad=0, h_pad=0)
# plt.savefig("continuous.png", dpi=300)
plt.legend()
plt.show()
plt.close()
```

### Central Limit Theorem

Imagine we draw N samples from some distribution and calculate the mean value, m. If we repeat this exercise many times, we will get a set of mean values, $\{m_0, m_1...\},$ each from a set of samples from the distribution. It doesn't matter if the N is the same each time, but N should't be too small $(>30)$

The central limit theorem states that the histogram or the probability distribution generated from this set of samples means, the $m$'s will approach a Gaussian in shape regardless of the shape of the distribution the samples were drawn from in the first place.


```python
M = 10000
m = np.zeros(M)
for i in range(M):
    t = np.random.beta(5, 2, size=M)
    m[i] = t.mean()
print("Mean of the means = %0.7f" % m.mean())

h, x = np.histogram(m, bins=B)
h = h / h.sum()
plt.bar(x[:-1] + 0.5 * (x[1] - x[0]), h, width=0.8 * (x[1] - x[0]))
plt.xlabel("Mean")
plt.ylabel("Probability")
plt.tight_layout(pad=0, w_pad=0, h_pad=0)
# plt.savefig("central_limit.png", dpi=300)
plt.show()
plt.close()
```

The above figure : The distribition of mean values of 10,000 set of samples of 10,000 from the Beta(5,2) distribution. The shape is evidently Gaussian with mean 0.7142891. There is a formula to calculate the mean value of a Beta distribution, mean of the Beta(5,2) = 0.714285. as you can see that both means are very close.

In summary, for any distribution, like in the example Beta(5,2), if we draw N samples, we can calculate the mean of those samples, a single number. If we repeat this process for many set of N samples, each of with its own mean from the same distribution and we make a histogram of the distribution of the means we measured, we will get a plot like the figure above. Which states that all of the many sample means are themselves clustered around a mean value. The mean value of the means is a measure of the population mean. It's the mean we would get if we could draw an infinite number of samples from the distribution.

### The Law of Large Numbers

The law of large numbers states that as the size of a sample from a distribution, the mean of the sample moves closer and closer to the mean of the population.

In this case, we are contemplating a single sample from the distribution and makeing a statement about how close we expect its mean to be to the true population mean. For central limit theorem, we have many different sets of samples from the distribution and are making a statement about the distribution of the means of those sets of sample.

we can demonstrate the law of large numbers quite simply by selecting larger and larger size samples from a distribution and tracking the mean as a function of the sample size


```python
m = []
for n in np.linspace(1, 8, 30):
    t = np.random.normal(1, 1, size=int(10**n))
    m.append(t.mean())

plt.plot(np.linspace(1, 8, 30), m)
plt.plot([1, 8], [1, 1], linestyle="--", color="k")
plt.xlabel("Exponent $10^n$")
plt.ylabel("Single sample mean")
plt.tight_layout(pad=0, w_pad=0, h_pad=0)
# plt.savefig("large_numbers.png", dpi=300)
plt.show()
```

we are drawing ever-larger sample size from a normal distribution with a mean of 1. The first sample size is 10 and last is 100 million. If we plot the mean of the samples as a fucntion of the sample size, we see the it gets closer and closer to the populuation mean of 1 thus proving the law of large numbers.
